```python
import pandas as pd
from sklearn import linear_model
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/5_one_hot_encoding/Exercise/carprices.csv")
df
```

```python
dummy=pd.get_dummies(df['Car Model'])
dummy
```

```python
merged=pd.concat([df,dummy],axis='columns')
merged
```

```python
final=merged.drop(['Car Model'], axis='columns')
final
```

```python
X=final.drop(['Sell Price($)'] , axis='columns')
X
```

```python
Y=final['Sell Price($)']
Y
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X,Y)
```

```python
model.predict([[45000,4,0,0,1]])
```

```python
model.score(X,Y)
```
